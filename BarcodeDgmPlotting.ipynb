```python
# Create numpy array from barcode string
def convert_bc(bc):
    # Input barcode as string of the form "[x1 y1; x2 y2; x3 y3]"
    # Converts to an nx2 numpy array for plotting barcodes
    bc_nobrack = bc[1:len(bc)-1]
    bc_split1 = bc_nobrack.split("; ")
    bc_arr = np.zeros((len(bc_split1), 2))
    #print(len(bc_split1))
    k=0
    for pair in bc_split1:
        #print(pair.split(" "))
        bc_arr[k,:] = pair.split(" ")
        k+=1
    return bc_arr
```
